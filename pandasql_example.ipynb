```python
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

from pandasql import load_meat, load_births
```

```python
# Load data
meat = load_meat()
births = load_births()
```

```python
# Pandas head
meat.head()
```

```python
# SQL Query
q = "SELECT * FROM meat LIMIT 5;"
df = pysqldf(q)
df
```

```python
# SQL join tables
q = """SELECT
        m.date, m.beef, b.births
     FROM
        meat m
     INNER JOIN
        births b
           ON m.date = b.date;"""
joined = pysqldf(q)
joined
```
